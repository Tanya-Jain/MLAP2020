```python
from scipy import stats
```

```python
stats.ttest_ind( autos[ autos.origin == 1]["mpg"],
              autos[ autos.origin == 3 ]["mpg"],
              equal_var=True)
```

```python
# outcome MPG - US vs JAPAN 
# average is different or not.
# alpha = 0.01
# ho: mean mPG for US = MEAN mPG for JAPAN.
# p value< alpha value so reject null hypothesis.
# p value is 4.17283714676552e-30
# inference: there is significant difference between mpg for vehicles manufactured by US and JAPAN.
```
